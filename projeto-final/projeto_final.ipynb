## Limpeza e Tratamento dos Dados

Antes de iniciar as análises, é necessário tratar os dados, lidando com valores nulos, ajustando tipos de dados e garantindo consistência nas informações.

### Tratamento de Valores Nulos

Removemos registros com valores ausentes nas colunas essenciais para a análise.

```python
# Removendo nulos nas tabelas principais
orders = orders.dropna()
items = items.dropna()
products = products.dropna()
customers = customers.dropna()
```

### Conversão de Tipos de Dados

Convertendo colunas de data para o formato adequado.

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

## Junção dos Dados

As tabelas serão combinadas para formar um único dataset consolidado, permitindo análises mais completas.

```python
df = items.merge(orders, on="order_id")
df = df.merge(products, on="product_id")
df = df.merge(customers, on="customer_id")
```

## Dataset Consolidado

Visualizamos o dataset final após a junção das tabelas.

```python
df.head()
```

```python
df.info()
```

## Criação de Variáveis

Criamos novas colunas que auxiliam na análise, como ano e mês da compra.

```python
df['ano'] = df['order_purchase_timestamp'].dt.year
df['mes'] = df['order_purchase_timestamp'].dt.month
```

## Verificação Final

Realizamos uma última verificação para garantir que os dados estão prontos para análise.

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

## Análise Exploratória e Visualizações

Nesta etapa, os dados serão analisados com base nas perguntas de negócio definidas anteriormente.

Cada análise será acompanhada de visualizações e interpretações, com o objetivo de extrair insights relevantes para o negócio.

### Evolução das Vendas ao Longo do Tempo

Nesta análise, buscamos entender como as vendas evoluíram ao longo dos anos, identificando possíveis tendências de crescimento ou queda.

```python
vendas_ano = df.groupby('ano')['price'].sum()

vendas_ano.plot(marker='o')
plt.title("Evolução das vendas por ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
plt.show()
```

**Interpretação:**

Observa-se a variação no volume de vendas ao longo dos anos. Essa análise permite identificar tendências de crescimento ou retração no negócio, sendo essencial para planejamento estratégico.

### Categorias de Produtos com Maior Receita

Esta análise identifica quais categorias de produtos são responsáveis pela maior parte da receita do e-commerce.

```python
top_categorias = (
    df.groupby('product_category_name')['price']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

sns.barplot(x=top_categorias.values, y=top_categorias.index)
plt.title("Top 10 categorias por receita")
plt.xlabel("Receita")
plt.ylabel("Categoria")
plt.show()
```

**Interpretação:**

Algumas categorias concentram grande parte da receita, indicando oportunidades de foco estratégico, سواء em marketing ou gestão de estoque.

### Distribuição de Receita por Estado

Nesta análise, buscamos entender a distribuição geográfica das vendas.

```python
receita_estado = (
    df.groupby('customer_state')['price']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

receita_estado.plot(kind='bar')
plt.title("Top estados por receita")
plt.xlabel("Estado")
plt.ylabel("Receita")
plt.show()
```

**Interpretação:**

A concentração de vendas em determinados estados pode indicar mercados mais fortes, auxiliando na definição de estratégias regionais.

### Ticket Médio dos Pedidos

O ticket médio representa o valor médio gasto por pedido, sendo um indicador importante do comportamento de consumo.

```python
ticket_medio = df.groupby('order_id')['price'].sum().mean()

ticket_medio
```

**Interpretação:**

O valor do ticket médio fornece insights sobre o padrão de consumo dos clientes, podendo indicar oportunidades de aumento de receita por meio de estratégias como upsell e cross-sell.

### Sazonalidade das Vendas

Analisamos o comportamento das vendas ao longo dos meses para identificar padrões sazonais.

```python
vendas_mes = df.groupby('mes')['price'].sum()

vendas_mes.plot(marker='o')
plt.title("Vendas por mês")
plt.xlabel("Mês")
plt.ylabel("Receita")
plt.show()
```

**Interpretação:**

A identificação de padrões sazonais permite antecipar períodos de alta e baixa demanda, contribuindo para planejamento de campanhas e estoque.
