## 5. Limpeza de Dados

Muitas vezes os dados possuem:
- Valores nulos
- Linhas duplicadas
- Dados que precisamos filtrar

Vamos ver algumas operações de limpeza.

```python
dados2 = {
    "Nome": ["Ana", "João", "Maria", "Pedro", "Ana"],
    "Idade": [23, 35, None, 40, 23],
    "Salario": [2500, 4000, 3000, None, 2500]
}

df2 = pd.DataFrame(dados2)
df2
```

**Valores nulos**

```python
df2.isnull().sum()
```

**Preenchendo valores nulos**

```python
df2["Idade"] = df2["Idade"].fillna(df2["Idade"].mean())
df2["Salario"] = df2["Salario"].fillna(df2["Salario"].mean())

df2
```

**Removendo duplicados**

```python
df2 = df2.drop_duplicates()
df2
```
