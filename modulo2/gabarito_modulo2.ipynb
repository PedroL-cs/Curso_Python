Exercício 1

```python
import pandas as pd

dados = {
    "Nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "Idade": [23, 35, 29, 41, 30],
    "Cidade": ["Fortaleza", "Recife", "Fortaleza", "Salvador", "Recife"],
    "Salario": [2500, 4000, 3200, 5000, 2800]
}

df = pd.DataFrame(dados)
print(df)
```

Exercício 3

```python
import numpy as np

dados_nulos = {
    "Nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "Idade": [23, np.nan, 29, np.nan, 30],
    "Salario": [2500, 4000, np.nan, 5000, np.nan]
}

df_nulos = pd.DataFrame(dados_nulos)

# Verificar valores nulos
df_nulos.isnull().sum()

# Preencher nulos com média
df_nulos["Idade"].fillna(df_nulos["Idade"].mean(), inplace=True)
df_nulos["Salario"].fillna(df_nulos["Salario"].mean(), inplace=True)

# Exibir DataFrame atualizado
print(df_nulos)
```

Exercício 4

```python
dados_dup = {
    "Nome": ["Ana", "Bruno", "Ana", "Carlos", "Carlos"],
    "Idade": [23, 35, 23, 29, 29],
    "Cidade": ["Fortaleza", "Recife", "Fortaleza", "Salvador", "Salvador"]
}

df_dup = pd.DataFrame(dados_dup)

# Exibir DataFrame original
print(df_dup)

# Remover duplicados
df_dup = df_dup.drop_duplicates()

# Exibir DataFrame sem duplicados
print(df_dup)
```

Exercício 8

```python
dados_alunos = {
    "Nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "Curso": ["ADS", "Engenharia", "ADS", "Direito", "ADS"],
    "Nota": [8.5, 6.0, 7.5, 9.0, 5.5],
    "Frequencia": [80, 75, 90, 95, 60]
}

alunos = pd.DataFrame(dados_alunos)

# Média das notas
alunos["Nota"].mean()

# Aluno com maior nota
alunos.loc[alunos["Nota"].idxmax()]

# Alunos com nota maior que 7
alunos[alunos["Nota"] > 7]

# Ordenar alunos pela nota
alunos.sort_values(by="Nota", ascending=False)
```
