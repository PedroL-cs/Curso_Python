# Curso de Python Introdutório
Este notebook contém os fundamentos da linguagem de forma organizada e comentada.

```python
# --- 1. Hello World ---
# A função print() exibe informações no console.
print("Olá, mundo! Bem-vindo ao curso de Python.")
```

```python
# --- 2. Variáveis e Tipagem (O Jeito Python de ser) ---
# TIPAGEM DINÂMICA: Não precisamos dizer o tipo. O Python descobre!
nome = "Fillipe"    # str (string)
idade = 25          # int (inteiro)
altura = 1.75       # float (ponto flutuante)
online = True       # bool (booleano)

# TIPAGEM FORTE: O Python exige que os tipos sejam compatíveis.
# Exemplo de conversão (Casting):
idade_texto = str(idade) # Transforma o número 25 no texto "25"
print("Tipo da idade agora: " + str(type(idade_texto)))

# TYPE HINTS (Opcional): Podemos apenas 'sugerir' o tipo para ajudar outros devs
hobby: str = "Programar"
print(f"{nome} tem {idade} anos e gosta de {hobby}.")
```

```python
# --- 3. Operações Aritméticas e Binárias ---
soma = 10 + 5
exponente = 2 ** 3    # 2 elevado a 3 (8)
resto = 10 % 3        # Resto da divisão (1)

# Operações Binárias (Bitwise) - Operam nos bits dos números
a = 5  # 101 em binário
b = 3  # 011 em binário
resultado_and = a & b # 001 em binário (Resultado: 1)
```

```python
# --- 4. Operações Lógicas (Estruturas de Decisão) ---
if idade >= 18:
    print(f"{nome} é maior de idade.")
else:
    print(f"{nome} é menor de idade.")

# Match (Python 3.10+) - Similar ao switch/case
status_code = 200
match status_code:
    case 200: print("Sucesso!")
    case 404: print("Não encontrado")
    case _: print("Erro desconhecido")
```

```python
# --- 5. Listas e Tuplas ---
frutas = ["maçã", "banana"] # Listas [] são mutáveis (podem mudar)
coordenadas = (10, 20)      # Tuplas () são imutáveis (fixas)

frutas.append("uva")        # Adicionando item à lista
```

```python
# --- 6. Dicionários ---
# Estruturas de Chave: Valor (Como um arquivo JSON)
aluno = {
    "nome": "João Silva",
    "curso": "Python Introdutório",
    "nota": 9.5
}

print(f"Aluno: {aluno['nome']} | Curso: {aluno['curso']}")

# Métodos úteis
print(f"Chaves: {aluno.keys()}")
print(f"Valores: {aluno.values()}")
```

```python
# --- 7. Laços de Repetição ---
# For: Percorre uma sequência
for fruta in frutas:
    print(f"Fruta da lista: {fruta}")

# Enumerate: Pega o índice e o valor ao mesmo tempo
for i, fruta in enumerate(frutas, start=1):
    print(f"{i}º lugar: {fruta}")

# While: Executa enquanto a condição for verdadeira
contador = 0
while contador < 3:
    print(f"Contagem While: {contador}")
    contador += 1
```

```python
# --- 8. Funções e Yields ---
# 'def' define um bloco reutilizável
def saudar(usuario: str) -> str:
    return f"Olá, {usuario}!"

print(saudar("Estudante"))

# Yield (Geradores): Retorna um valor por vez, pausando a função
def contador_passo_a_passo():
    yield "Passo 1"
    yield "Passo 2"
    yield "Passo 3"

for passo in contador_passo_a_passo():
    print(passo)
```

```python
# --- 9. Importação de Módulos ---
# Importamos ferramentas prontas antes de começar o código principal.
# Importar o módulo inteiro
import math
print(math.sqrt(25))  # Raiz quadrada de 25

# Importar algo específico (mais limpo)
from datetime import datetime
print(datetime.now()) # Exibe a data e hora atual

# Importar com um "apelido" (muito usado em Data Science)
import random as rd
numero_sorteado = rd.randint(1, 10) # Gera número aleatório entre 1 e 10
```

```python
# --- 10. Classes (Orientação a Objetos) ---
class Robo:
    def __init__(self, nome: str):
        self.nome = nome  # Atributo
    
    def falar(self):      # Método (Ação)
        print(f"{self.nome} diz: Beep Boop! Sistema operacional ativo.")

meu_robo = Robo("R2-D2")
meu_robo.falar()
```
