# Análise Exploratória de Dados – Dataset Tips

Neste projeto será realizada uma Análise Exploratória de Dados (EDA) utilizando o dataset Tips.
O objetivo é analisar o comportamento das gorjetas em um restaurante, buscando entender
como fatores como valor da conta, dia da semana, período do dia e número de pessoas influenciam
o valor da gorjeta.

## Problema

Um restaurante deseja entender o comportamento das gorjetas dos clientes.
A empresa quer identificar padrões e fatores que influenciam o valor das gorjetas,
como valor da conta, dia da semana, horário e número de pessoas na mesa.

## Perguntas Norteadoras

1. Qual é o valor médio das gorjetas?
2. Gorjetas maiores estão associadas a contas maiores?
3. Qual dia da semana possui maior média de gorjetas?
4. O jantar possui gorjetas maiores que o almoço?
5. O número de pessoas influencia o valor da gorjeta?
6. Existe relação entre o sexo do cliente e o valor da gorjeta?

---

## 1. Importação das Bibliotecas

Para realizar a análise de dados, precisamos importar algumas bibliotecas do Python:

- pandas: manipulação e análise de dados
- matplotlib: criação de gráficos
- seaborn: visualização de dados mais avançada

Essas bibliotecas são amplamente utilizadas em projetos de análise de dados e ciência de dados.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---

## 2. Carregamento do Dataset

Nesta etapa vamos carregar o conjunto de dados que será analisado.
O dataset utilizado se chama "Tips" e contém informações sobre contas
de um restaurante, incluindo valor da conta, gorjeta, dia da semana,
período (almoço ou jantar), sexo do cliente e número de pessoas na mesa.

Após carregar os dados, vamos visualizar as primeiras linhas do dataset
para entender como os dados estão organizados.

```python
tips = sns.load_dataset("tips")
tips.head()
```

---

## 3. Inspeção Inicial dos Dados

Antes de iniciar a análise, é importante entender a estrutura do dataset.
Vamos verificar:

- Quantas linhas e colunas existem
- Tipos de dados
- Estatísticas básicas
- Valores nulos

Isso ajuda a identificar possíveis problemas nos dados.

```python
tips.info()
tips.describe()
tips.isnull().sum()
```

### Observações

A partir das informações acima, podemos observar a estrutura do dataset,
os tipos de dados presentes e algumas estatísticas iniciais como média,
mínimo, máximo e desvio padrão.

Também verificamos se existem valores nulos que precisariam ser tratados
antes da análise.

---

## 4. Análise Estatística

Agora vamos calcular algumas estatísticas importantes para entender melhor
o comportamento das gorjetas e das contas do restaurante.

Vamos analisar:
- Média das gorjetas
- Média das contas
- Gorjeta média por dia da semana
- Gorjeta média por período (almoço ou jantar)
- Gorjeta média por número de pessoas

```python
# Média das gorjetas
tips["tip"].mean()

# Média da conta
tips["total_bill"].mean()

# Gorjeta média por dia
tips.groupby("day")["tip"].mean()

# Gorjeta média por período
tips.groupby("time")["tip"].mean()

# Gorjeta média por número de pessoas
tips.groupby("size")["tip"].mean()
```

### Interpretação

Com essas estatísticas podemos comparar os valores médios entre diferentes
categorias e identificar padrões iniciais nos dados, como quais dias possuem
maiores gorjetas ou se mesas com mais pessoas geram gorjetas maiores.

---

## 5. Visualizações de Dados

A visualização de dados é uma etapa muito importante da análise exploratória,
pois permite identificar padrões, tendências e possíveis outliers de forma
mais clara através de gráficos.

A seguir, serão apresentados alguns gráficos para analisar o comportamento
das contas e gorjetas.

### Histograma – Contas

```python
plt.hist(tips["total_bill"])
plt.title("Distribuição do Valor das Contas")
plt.xlabel("Valor da Conta")
plt.ylabel("Frequência")
plt.show()
```

### Histograma – Gorjetas

```python
plt.hist(tips["tip"])
plt.title("Distribuição das Gorjetas")
plt.show()
```

### Scatter – Conta vs Gorjeta

```python
plt.scatter(tips["total_bill"], tips["tip"])
plt.xlabel("Conta")
plt.ylabel("Gorjeta")
plt.title("Conta vs Gorjeta")
plt.show()
```

### Boxplot – Gorjeta por Dia

```python
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Gorjeta por Dia da Semana")
plt.show()
```

### Barplot – Gorjeta média por Dia

```python
sns.barplot(x="day", y="tip", data=tips)
plt.title("Média de Gorjeta por Dia")
plt.show()
```

### Barplot – Gorjeta por Período

```python
sns.barplot(x="time", y="tip", data=tips)
plt.title("Gorjeta por Período (Almoço/Jantar)")
plt.show()
```

### Barplot – Gorjeta por Sexo

```python
sns.barplot(x="sex", y="tip", data=tips)
plt.title("Gorjeta por Sexo")
plt.show()
```

---

## 6. Insights

Após a análise exploratória e a visualização dos dados, foi possível identificar
alguns padrões importantes no comportamento das contas e gorjetas do restaurante.

### Distribuição das Contas
Observando o histograma do valor das contas, percebemos que a maioria das contas
está concentrada em valores mais baixos, enquanto poucas contas possuem valores
muito altos, caracterizando possíveis outliers.

### Distribuição das Gorjetas
A distribuição das gorjetas segue um padrão semelhante ao das contas, com a maioria
das gorjetas em valores menores e poucas gorjetas muito altas.

### Relação entre Conta e Gorjeta
No gráfico de dispersão (scatter plot), foi possível observar uma relação positiva
entre o valor da conta e o valor da gorjeta. Isso indica que, em geral, quanto maior
a conta, maior a gorjeta deixada pelo cliente.

### Gorjeta por Dia da Semana
Através do boxplot e do barplot por dia da semana, observamos que alguns dias
apresentam maior média de gorjetas, indicando que o comportamento dos clientes
pode variar dependendo do dia.

### Gorjeta por Período (Almoço ou Jantar)
A análise mostrou que o período do jantar geralmente apresenta gorjetas maiores
do que o período do almoço, possivelmente porque as contas do jantar costumam ser maiores.

### Gorjeta por Sexo
Comparando as gorjetas por sexo, observamos pequenas diferenças nos valores médios,
mas sem uma diferença muito significativa entre homens e mulheres.

### Correlação entre Variáveis
A matriz de correlação mostrou que a variável que possui maior relação com a gorjeta
é o valor da conta total, reforçando a ideia de que a gorjeta está diretamente
relacionada ao valor gasto pelo cliente.

### Insight Geral
De forma geral, o fator que mais influencia o valor da gorjeta é o valor da conta,
enquanto fatores como dia da semana, período e número de pessoas também possuem
influência, mas em menor escala.
