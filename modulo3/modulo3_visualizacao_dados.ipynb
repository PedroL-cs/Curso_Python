## 1. Introdução à Visualização de Dados

Visualização de dados é a etapa em que transformamos números, tabelas e estatísticas em gráficos e representações visuais.

Ela é importante porque ajuda a:

- identificar padrões;
- comparar grupos;
- encontrar relações entre variáveis;
- detectar valores fora do padrão (outliers);
- comunicar descobertas com mais clareza.

Em Ciência de Dados, não basta analisar bem — também é preciso explicar os resultados de forma compreensível.

Exemplo:

Uma tabela pode mostrar várias notas de alunos, mas um gráfico pode revelar rapidamente:

qual turma teve melhor desempenho;
se há grande variação entre alunos;
se existe relação entre estudo e nota.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Para carregar a base de dados usada nos exemplos você utiliza o seguinte comando:

df = pd.read_csv("dataset_desempenho_escolar.csv")
df.head()
```

---

## 2. Visualização com Matplotlib

O Matplotlib é uma das bibliotecas mais utilizadas em Python para criação de gráficos.

Ele permite construir gráficos do zero e controlar detalhes como:

- tamanho da figura;
- títulos;
- rótulos dos eixos;
- legendas;
- grades;
- estilo visual.

### 2.1. Gráfico de Barras

O gráfico de barras é ideal para comparar categorias.

Exemplo: comparar a média das notas entre turmas.

```python
media_matematica = df.groupby("turma")["nota_matematica"].mean()

plt.bar(media_matematica.index, media_matematica.values)
plt.title("Média de Matemática por Turma")
plt.xlabel("Turma")
plt.ylabel("Média")
plt.show()
```

Quando usar:
- comparação entre grupos;
- ranking;
- contagem de categorias.

### 2.2. Histograma

O histograma mostra como os valores de uma variável estão distribuídos.

Ele ajuda a responder perguntas como:

- As notas estão concentradas em valores altos ou baixos?
- Existe grande variação?
- Há muitos valores extremos?

```python
plt.hist(df["nota_matematica"], bins=10)
plt.title("Distribuição das Notas de Matemática")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.show()
```

Quando usar:
- distribuição de dados numéricos;
- frequência de valores;
- análise de dispersão.

### 2.3. Gráfico de Dispersão (Scatter Plot)

O gráfico de dispersão mostra a relação entre duas variáveis numéricas.

Ele é útil para investigar se existe alguma associação entre elas.

```python
plt.scatter(df["horas_estudo"], df["nota_matematica"])
plt.title("Horas de Estudo x Nota de Matemática")
plt.xlabel("Horas de Estudo")
plt.ylabel("Nota de Matemática")
plt.show()
```

Quando usar:
- investigar correlação;
- observar tendências;
- detectar agrupamentos ou outliers.

### 2.4. Customização Básica

Um gráfico bom não é apenas um gráfico que funciona.
Ele precisa ser legível, organizado e claro.

Veja um exemplo com melhorias visuais:

```python
plt.figure(figsize=(8, 5))
plt.bar(media_matematica.index, media_matematica.values)
plt.title("Média de Matemática por Turma", fontsize=14)
plt.xlabel("Turma")
plt.ylabel("Média")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

Elementos importantes:
- Título
- Nome dos eixos
- Escala adequada
- Tamanho da figura
- Grid quando útil

## 3. Visualização com Seaborn

O Seaborn é uma biblioteca construída sobre o Matplotlib, mas com foco em gráficos mais elegantes e mais adequados para análise estatística.

Ele facilita bastante a criação de visualizações bonitas com pouco código.

### 3.1. Boxplot

O boxplot resume a distribuição dos dados usando:

- mínimo;
- primeiro quartil;
- mediana;
- terceiro quartil;
- máximo;
- outliers.

```python
sns.boxplot(x="turma", y="nota_matematica", data=df)
plt.title("Boxplot das Notas de Matemática por Turma")
plt.show()
```

Quando usar:
- comparar distribuição entre grupos;
- detectar outliers;
- analisar dispersão.

### 3.2. Violin Plot

O violin plot é semelhante ao boxplot, mas também mostra a densidade da distribuição.

Ele ajuda a enxergar onde os dados estão mais concentrados.

```python
sns.violinplot(x="turma", y="nota_matematica", data=df)
plt.title("Distribuição das Notas de Matemática por Turma")
plt.show()
```

Quando usar:
- comparar distribuições com mais detalhe;
- visualizar concentração dos dados.

### 3.3. Heatmap

O heatmap é muito usado para mostrar correlações entre variáveis numéricas.

```python
correlacao = df[["nota_matematica", "nota_portugues", "frequencia", "horas_estudo"]].corr()

sns.heatmap(correlacao, annot=True, cmap="Blues")
plt.title("Mapa de Correlação")
plt.show()
```

Quando usar:
- comparar várias variáveis ao mesmo tempo;
- encontrar relações entre colunas;
- visualizar padrões rapidamente.

### 3.4. Regressão Linear

O gráfico de regressão é uma forma de observar se existe uma tendência linear entre duas variáveis.

```python
sns.regplot(x="horas_estudo", y="nota_matematica", data=df)
plt.title("Horas de Estudo x Nota de Matemática")
plt.show()
```

Se a linha sobe, isso pode indicar que, conforme uma variável aumenta, a outra também tende a aumentar.

Quando usar:
- investigar tendência;
- observar relação entre duas variáveis;
- introduzir análise preditiva básica.
