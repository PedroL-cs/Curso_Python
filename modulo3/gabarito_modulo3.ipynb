Exercício 2

```python
import matplotlib.pyplot as plt

meses = [1,2,3,4,5,6]
vendas = [100, 120, 150, 130, 170, 200]

plt.plot(meses, vendas, marker='o')
plt.title("Vendas por Mês")
plt.xlabel("Meses")
plt.ylabel("Vendas")
plt.show()
```

Exercício 5

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dados = {
    "Cidade": ["Fortaleza", "Recife", "Salvador", "Fortaleza", "Recife", "Salvador"],
    "Vendas": [200, 150, 180, 220, 170, 190]
}

df = pd.DataFrame(dados)

sns.barplot(x="Cidade", y="Vendas", data=df)
plt.title("Média de Vendas por Cidade")
plt.show()
```

Exercício 6

```python
import seaborn as sns
import matplotlib.pyplot as plt

salarios = [2000, 2200, 2500, 2700, 3000, 3200, 3500, 4000, 4500, 10000]

sns.boxplot(x=salarios)
plt.title("Boxplot dos Salários")
plt.show()
```

Respostas:

1. Sim, 10000 é outlier
2. Mediana ~ 3000
3. Mostra distribuição, mediana e outliers

Exercício 7

```python
import matplotlib.pyplot as plt

anos = [2018,2019,2020,2021,2022,2023]
lucro = [10, 15, 8, 20, 25, 30]

plt.plot(anos, lucro, color='red', marker='o')
plt.title("Lucro por Ano")
plt.xlabel("Ano")
plt.ylabel("Lucro")
plt.grid(True)
plt.show()
```
